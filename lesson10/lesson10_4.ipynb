```python
import csv
from pprint import pprint
with open('abc.csv', encoding='utf-8',newline='') as file:
    #content = file.read()
    dictReader = csv.DictReader(file)
    contents:list = list(dictReader)

pprint(contents) #List of Dictionary
#總之看起來就是比前一個舒服一點
```

```python
for item in contents:
    if '元大' in item['證券名稱']:
        print(item)
#先召喚上面的資料再用這一串
```

```python
searchs:list = []
for item in contents:
    if '元大' in item['證券名稱']:
        searchs.append(item)

pprint(searchs)
```
