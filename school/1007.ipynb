```python
#輸出個人期末考成績單，計算總分及平均，找出表現最好和最差的科目，比較兩次段考成績

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#成績資料
data = {
    "科目": ["國文", "英文", "數學", "社會", "自然"],
    "期中考分數": [85, 70, 92, 88, 75],
    "期末考分數": [72, 60, 100, 80, 54]
}

# 用 Pandas 建立 DataFrame
df = pd.DataFrame(data)
df_final = df[['科目', '期末考分數']]

#計算
scores = np.array(df['期末考分數'])
scores_sum = np.sum(scores)
scores_average = np.mean(scores)
scores_max = df.loc[df["期末考分數"].idxmax(), "科目"]
scores_min = df.loc[df["期末考分數"].idxmin(), "科目"]

#提取
scores_midterm = np.array(df['期中考分數'])
subjects = df['科目']

#畫圖
plt.plot(subjects, scores, marker='o', color='red')
plt.plot(subjects, scores_midterm, marker='o', color='blue')
plt.title("段考成績比較")
plt.xlabel("subjects")
plt.ylabel("scores")
plt.ylim(0, 110) #設定y軸範圍

#輸出
print("期末考成績單")
print(df_final)
print("\n總分：", scores_sum)
print("平均：", scores_average)
print("表現最好的科目是：", scores_max)
print("需要加強的科目是：", scores_min)
print("\n本學期段考成績比較")
plt.show()
```
